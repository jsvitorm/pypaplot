```python
import sqlite3

import pandas as pd

import requests

import matplotlib.pyplot as plt

conn = sqlite3.connect('banco.db')

c = conn.cursor()
```

```python
lista_pokemons = ["charmander", "ditto", "squirtle","pikachu","gastly", "zoroark"]

chaves = ["nome_poke", "peso", "altura"]
valores = []

dicionario_pokemons = {}

cont = 0
for pokemon in lista_pokemons:
    req = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon}/')
    data = req.json()
    nome = data['name']
    peso = data['weight']
    altura = data['height']
    dici = {
        'nome_poke':nome,
        'peso':peso,
        'altura':altura
    }
    dicionario_pokemons[cont] = dici
    cont=cont+1

dicionario_pokemons

```

```python
dicionario_pokemons
```

```python
df_poke = pd.DataFrame.from_dict(dicionario_pokemons, orient='index')

df_poke
```

```python
fig, ax = plt.subplots()


df_poke['altura'].plot(kind="bar", ax=ax)

ax.set_xticklabels(df_poke["nome_poke"])


plt.show()
```

```python
df_poke.to_sql('pokemons',conn)
```
